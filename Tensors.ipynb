## Inplace operation

```python
#so what happens is that whenever we perform some operations it stores that in a new tensor, which occupies the space so to reduce that we store that in the previous tensor 
```

```python
m=torch.rand(2,3)
n=torch.rand(2,3)
print(m)
print(n)
```

```python
m.add_(n)
```

```python
m
```
